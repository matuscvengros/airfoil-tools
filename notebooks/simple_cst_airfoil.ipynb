# Parametrisation of a CST Airfoil
This notebook follows an example of simple parametrisation of a Class-Shape Transform (CST) airfoil.

Following cell introduces the project folder into system path. It's a hackish for the Jupyter Kernal to understand where is the `airfoiltools` package located without the need to force a static or symbolic install.

```python
from pathlib import Path
import sys

PROJECT_SRC = Path().cwd().parent
sys.path.append(str(PROJECT_SRC))

%load_ext autoreload
%autoreload 2
```

```python
import airfoiltools as aft
import matplotlib.pyplot as plt

airfoil = aft.CST([0.26, 0.21, 0.52, 0.09, 0.67, 0.14, 0.29, 0.16], [-0.26, -0.21, -0.11, -0.07, -0.17, -0.15, 0.12, 0.02], 0.54, 0.01, name="CST Sample Airfoil", chord=250)


plt.figure(figsize=(10, 4))
plt.plot(airfoil.psi, airfoil.upper, label='Upper Surface', color='purple')
plt.plot(airfoil.psi, airfoil.lower, label='Lower Surface', color='orange')
plt.title(f'{airfoil.name}')
plt.xlabel(r'$\bar{x}$')
plt.ylabel(r'$\bar{y}$')
plt.axis('equal')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(airfoil.coords[:, 0], airfoil.coords[:, 1], color='black')
plt.title(f'Airfoil Shape')
plt.xlabel('x [mm]')
plt.ylabel('y [mm]')
plt.axis('equal')
plt.grid(True)
plt.show()
```
